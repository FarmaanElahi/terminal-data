```python
import pandas as pd
import yfinance as yf
import numpy as np

# Step 1: 20 dummy stock tickers
tickers = [
    'AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'NVDA', 'META', 'BRK-B', 'JNJ', 'V',
    'PG', 'JPM', 'UNH', 'MA', 'HD', 'KO', 'PEP', 'LLY', 'MCD', 'DIS'
]

# Step 2: Download past 400 days of data
data = yf.download(tickers, period='400d')['Close'].dropna()

# Step 3: Lookback periods (days)
lookbacks = {
    '1M': 21,
    '3M': 63,
    '6M': 126,
    '12M': 252
}

# Step 4: Calculate % change for each period
pct_changes = pd.DataFrame(index=data.columns)

for label, days in lookbacks.items():
    if len(data) >= days:
        pct_change = ((data.iloc[-1] - data.iloc[-days]) / data.iloc[-days]) * 100
        pct_changes[label] = pct_change.round(2)

# Step 5: Rank based on percentile
as_ratings = pd.DataFrame(index=data.columns)

for label in lookbacks.keys():
    if label in pct_changes:
        # Percentile rank: higher % change = higher rating
        percentiles = pct_changes[label].rank(pct=True, ascending=True) * 100
        as_ratings[f'{label}_AS_Rating'] = percentiles.round(2)

print("\nAbsolute Strength Ratings:\n")
as_ratings
```
